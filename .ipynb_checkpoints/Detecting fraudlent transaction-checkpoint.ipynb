```python
# Import libraries
import os, sys
import pandas as pd
import numpy as np
import seaborn as sns
sns.set()
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv(r"C:\Users\91866\Downloads\payment_fraud.csv") # Load the dataset
```

```python
df.info() # Basic information about the dataset
# We have 7 independent variable and 1 depended variable
# Its a classification problem
# There are some missing values in dataset
```

```python
df.describe()
# There are accounts which are in range of 0-2000 days old
# The recency of the payment transaction initiated from range 0-2000 days old
# numitems: There are other items associated with the account which might include Fixed deposits, recurring deposits,
#           Investment services etc. upto 29 other such items linked to the account.
# Label is the dependent variable which consist of two output 1 and 0 whether the transaction is fraudlent or not. 
#   (binary classification)
```

```python
print('*'*20,'IsWeekend','*'*20)
print(df['isWeekend'].value_counts())
```

```python
# All the 560 missing values for isWeekend belongs to fraudulent transactions.
df[df['label']==1].loc[:,'isWeekend'].isnull().sum()
```

```python
# We will delete the isWeekend feature from the dataset
# isWeekend - There are null vlaues for all the observations which are fraudulent
df.drop(['isWeekend'],axis=1,inplace=True)
```

```python
# Imputing the missing values with mode value of category.
df['Category'].fillna(df['Category'].mode()[0],inplace=True)
```

```python
df.isnull().sum()
# Now there are no more missing values in the data
```

```python
# Converting the local time into datetime format to extract the time of the transaction
df['formated_date'] = pd.to_datetime(df['localTime'],unit='d')
```

```python
# Converting the date into string and extracting time out of it, excluding the date
df['formated_date'] = df['formated_date'].astype(str)
df[['date','time']] = df['formated_date'].str.split(' ',1,expand=True)
```

```python
df.drop(['date','localTime','formated_date'],axis=1,inplace=True)
```

```python
# Converting the time feature into correct time format
df['time'] = pd.to_datetime(df['time'],format='%H:%M:%S.%f').dt.strftime('%H:%M:%S')
```

```python
# Function to identify whether its a day or night
def categorize_time(time):
    hour = int(time.split(':')[0])
    if hour <=18:
        return 'Day'
    return 'Night'
```

```python
df['time_category'] = df['time'].apply(categorize_time)
```

```python
# Deleting the time and numitems column from the dataset
df_1 = df.drop(['time','numItems'],axis=1)
```

```python
# Splitting the data into independent and dependent variable
x = df_1.drop(['label'],axis=1)
y = df_1['label']
```

```python
# Outlier treatment 
Q1, Q3 = x['paymentMethodAgeDays'].quantile((0.25,0.75))
IQR = Q3-Q1
upper_range = Q3 + 1.5*IQR
x['paymentMethodAgeDays'] = np.where(x['paymentMethodAgeDays']>Q3,upper_range,x['paymentMethodAgeDays'])
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve, classification_report,precision_score,recall_score
```

```python
# Scaling and one-hot encoding the dataset
num_variables = x[['accountAgeDays','paymentMethodAgeDays']]
cat_variables = x[['paymentMethod','Category','time_category']]
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
x_scaled = scaler.fit_transform(num_variables)
x_scaled = pd.DataFrame(x_scaled,columns=num_variables.columns)
cat_variables = pd.get_dummies(cat_variables,drop_first=True)
x = pd.concat([x_scaled,cat_variables],axis=1)
```

```python
x.head()
```

```python
# Splitting the date into train test split
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,y,stratify=y,random_state=101)
```

```python
model_evaluation = pd.DataFrame(columns=['Model name', 'Train accuracy', 'Test accuracy', 'Precision', 'Recall', 
                                         'Dataset type'])
```

```python
# Training the model with Logistic regression
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(x_train,y_train)
y_test_predict = model.predict(x_test)
y_train_predict = model.predict(x_train)
model_test_accuracy_score = accuracy_score(y_test,y_test_predict)
model_train_accuracy_score = accuracy_score(y_train,y_train_predict)
print(classification_report(y_test,y_test_predict))
model_precision_score = precision_score(y_test,y_test_predict)
model_recall_score = recall_score(y_test,y_test_predict)
sns.heatmap(confusion_matrix(y_test,y_test_predict),annot=True,fmt='d')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.title('Confusion matrix')
current_model = pd.DataFrame({'Model name': [model.__class__.__name__],
                              'Train accuracy': [model_train_accuracy_score],
                              'Test accuracy': [model_test_accuracy_score],
                              'Precision': [model_precision_score],
                              'Recall': [model_recall_score],
                              'Dataset type': ['Imbalanced dataset']})
model_evaluation = model_evaluation.append(current_model, ignore_index=True)
model_evaluation

# Overfitting issues because no equal number of samples provided in the model to train 
```

```python
# Training the model with KNN regression
from sklearn.svm import SVC
model = SVC()
model.fit(x_train,y_train)
y_test_predict = model.predict(x_test)
y_train_predict = model.predict(x_train)
model_test_accuracy_score = accuracy_score(y_test,y_test_predict)
model_train_accuracy_score = accuracy_score(y_train,y_train_predict)
print(classification_report(y_test,y_test_predict))
model_precision_score = precision_score(y_test,y_test_predict)
model_recall_score = recall_score(y_test,y_test_predict)
sns.heatmap(confusion_matrix(y_test,y_test_predict),annot=True,fmt='d')
plt.xlabel('Predicted values')
plt.ylabel('Actual values')
plt.title('Confusion matrix')
current_model = pd.DataFrame({'Model name': [model.__class__.__name__],
                              'Train accuracy': [model_train_accuracy_score],
                              'Test accuracy': [model_test_accuracy_score],
                              'Precision': [model_precision_score],
                              'Recall': [model_recall_score],
                              'Dataset type': ['Imbalanced dataset']})
model_evaluation = model_evaluation.append(current_model, ignore_index=True)
model_evaluation
# Overfitting issues because no equal number of samples provided in the model to train 
```
